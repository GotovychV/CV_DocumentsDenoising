```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [14, 10]
```

```python
# Set the seeds for reproducibility
from numpy.random import seed
from tensorflow.random import set_seed
seed_value = 1234578790
seed(seed_value)
set_seed(seed_value)
```

```python
dataFolder = 'Data'
```

```python
for file in os.listdir(os.path.join(dataFolder, 'train')):
  img = cv2.imread(os.path.join(dataFolder, 'train', file))
  print(file + ' ' + str(img.shape))
```

```python
for file in os.listdir(os.path.join(dataFolder, 'train_cleaned')):
  img = cv2.imread(os.path.join(dataFolder, 'train_cleaned', file))
  print(file + ' ' + str(img.shape))
```

```python
def GetData(imgCount, basicFolder, subFolder):
  data = np.empty((imgCount, 420, 540, 3), dtype=int)

  k = 0
  for file in os.listdir(os.path.join(basicFolder, subFolder)):
    img = cv2.imread(os.path.join(basicFolder, subFolder, file))
    if img.shape[0] == 420:
      data[k,:,:,:] = img
    else:
      img = cv2.resize(img, (540, 420))
      data[k,:,:,:] = img
    k += 1

  return data
```

```python
trainData = GetData(144, dataFolder, 'train')
```

```python
trainDataCleaned = GetData(144, dataFolder, 'train_cleaned')
```

```python
counter = 1
for ii in range(4):
  imgNumber = np.random.randint(0, 144)
  plt.subplot(4, 2, counter), plt.imshow(trainData[imgNumber,:,:,:])
  plt.subplot(4, 2, counter + 1), plt.imshow(trainDataCleaned[imgNumber,:,:,:])
  counter += 2
```

```python
testData = GetData(72, dataFolder, 'test')
```

```python
for ii in range(4):
  imgNumber = np.random.randint(0, 72)
  plt.subplot(4, 2, ii + 1), plt.imshow(testData[imgNumber,:,:,:])
```
